# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

# Calculate the date 1 year ago from the last data point in the database
measurements = session.query(Measurement)

last_date = session.query(func.max(Measurement.date))
last_date_list=[date[0] for date in last_date]
last_date_date=dt.datetime.strptime(last_date_list[0],'%Y-%m-%d')
twelve_months_ago=last_date_date-dt.timedelta(days=365)
twelve_months_ago_string=twelve_months_ago.strftime('%Y-%m-%d')

# Perform a query to retrieve the data and precipitation scores
prec_scores = session.query(Measurement.prcp,Measurement.date).filter(Measurement.date>twelve_months_ago_string)

prec_list=[prec[0] for prec in prec_scores]
date_list=[prec[1] for prec in prec_scores]

# Save the query results as a Pandas DataFrame and set the index to the date column

#prec_df = pd.DataFrame(prec_list,index=date_list,columns=['precipitation']).dropna()
prec_df = pd.DataFrame(list(zip(prec_list,date_list)),columns=['precipitation','date']).dropna()

# Sort the dataframe by date

prec_df.sort_index()

#prec_df.head()
# Use Pandas Plotting with Matplotlib to plot the data

fig=plt.figure(figsize=(20,10))
plt.bar(prec_df.date,prec_df.precipitation)
plt.xticks(prec_df.date, [])
plt.title('Precipitation in Last Twelve Months',fontsize=30)
plt.xlabel('Date',fontsize=25)
plt.ylabel('Precipitation',fontsize=25)
plt.show()
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
```

```python
stations=session.query(Station.station)
station_list=[station[0] for station in stations]
activity_count = []
for station in station_list:
    activity_count.append(session.query(Measurement.prcp).filter(Measurement.station==station).count())

stations_df = pd.DataFrame(list(zip(station_list,activity_count)),columns=['station','activity']).sort_values(by='activity',ascending=False)
stations_df
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
lowest_temp_active_station=session.query(func.min(Measurement.tobs)).filter(Measurement.station==stations_df.station[6])
highest_temp_active_station=session.query(func.max(Measurement.tobs)).filter(Measurement.station==stations_df.station[6])
avg_temp_active_station=session.query(func.avg(Measurement.tobs)).filter(Measurement.station==stations_df.station[6])
station_stat_list_lowest=[lowest_temp_active_station[0] for temp in lowest_temp_active_station]
station_stat_list_highest=[highest_temp_active_station[0] for temp in highest_temp_active_station]
station_stat_list_avg = [avg_temp_active_station[0] for temp in avg_temp_active_station]

data=[[station_stat_list_lowest[0][0],station_stat_list_highest[0][0],station_stat_list_avg[0][0]]]
station_stats_df = pd.DataFrame(data, columns = ['Lowest Temp', 'Highest Temp','Average Temp']) 
station_stats_df=station_stats_df.round(1)
station_stats_df

```

```python
# Choose the station with the highest number of temperature observations.
temp_activity_count = []
for station in station_list:
    temp_activity_count.append(session.query(Measurement.tobs).filter(Measurement.station==station).count())
stations_temp_df = pd.DataFrame(list(zip(station_list,temp_activity_count)),columns=['station','temp activity']).sort_values(by='temp activity',ascending=False)
stations_temp_df
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

temp_last_12 = session.query(Measurement.tobs,Measurement.date).filter(Measurement.date>twelve_months_ago_string).filter(Measurement.station==stations_df.station[6])
temp_last_12_df = pd.DataFrame(temp_last_12)
fig=plt.figure(figsize=(10,10))
plt.hist(temp_last_12_df.tobs,bins=12)
plt.ylabel('Frequency',fontsize=15)
plt.legend('tobs')
plt.show
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```
